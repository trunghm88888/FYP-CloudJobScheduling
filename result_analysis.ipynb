```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cpu_usage = pd.read_csv('logging/cpu_usage_24_03_2024_18_49_22.csv')
base_cpu_usage = pd.read_csv('logging/cpu_usage_24_03_2024_19_42_09.csv')
memory_usage = pd.read_csv('logging/memory_usage_24_03_2024_18_52_05.csv')
base_memory_usage = pd.read_csv('logging/memory_usage_24_03_2024_20_08_54.csv')
```

```python
all_cpu_usage = pd.DataFrame()
all_cpu_usage['time'] = cpu_usage['time']
all_cpu_usage['base_cpu_usage'] = base_cpu_usage['cpu_usage']
all_cpu_usage['stratus_cpu_usage'] = cpu_usage['cpu_usage']
all_cpu_usage
```

```python
plt.figure(figsize=(30, 10))
plt.rc('font', size=20)
sns.lineplot(data=all_cpu_usage, x='time', y='base_cpu_usage', label='Base CPU usage')
sns.lineplot(data=all_cpu_usage, x='time', y='stratus_cpu_usage', label='Stratus CPU usage')
plt.xlabel('Time (hours)')
plt.ylabel('Total acquired CPU resources (Unit)')
```

```python
all_cpu_utilization = pd.DataFrame()
all_cpu_utilization['time'] = cpu_usage['time']
all_cpu_utilization['base_cpu_utilization'] = base_cpu_usage['cpu_utilization']
all_cpu_utilization['stratus_cpu_utilization'] = cpu_usage['cpu_utilization']
all_cpu_utilization
```

```python
plt.figure(figsize=(30, 10))
plt.rc('font', size=20)
sns.lineplot(data=all_cpu_utilization, x='time', y='base_cpu_utilization', label='Base CPU utilization')
sns.lineplot(data=all_cpu_utilization, x='time', y='stratus_cpu_utilization', label='Stratus CPU utilization')
plt.xlabel('Time (hours)')
plt.ylabel('CPU utilization (%)')
```

```python
all_memory_usage = pd.DataFrame()
all_memory_usage['time'] = memory_usage['time']
all_memory_usage['base_memory_usage'] = base_memory_usage['memory_usage']
all_memory_usage['stratus_memory_usage'] = memory_usage['memory_usage']
all_memory_usage
```

```python
plt.figure(figsize=(30, 10))
plt.rc('font', size=20)
sns.lineplot(data=all_memory_usage, x='time', y='base_memory_usage', label='Base memory usage')
sns.lineplot(data=all_memory_usage, x='time', y='stratus_memory_usage', label='Stratus memory usage')
plt.xlabel('Time (hours)')
plt.ylabel('Total acquired memory resources (Unit)')
```

```python
all_memory_utilization = pd.DataFrame()
all_memory_utilization['time'] = memory_usage['time']
all_memory_utilization['base_memory_utilization'] = base_memory_usage['memory_utilization']
all_memory_utilization['stratus_memory_utilization'] = memory_usage['memory_utilization']
all_memory_utilization
```

```python
plt.figure(figsize=(30, 10))
plt.rc('font', size=20)
sns.lineplot(data=all_memory_utilization, x='time', y='base_memory_utilization', label='Base memory utilization')
sns.lineplot(data=all_memory_utilization, x='time', y='stratus_memory_utilization', label='Stratus memory utilization')
plt.xlabel('Time (hours)')
plt.ylabel('Memory utilization (%)')
```

```python
avg_cpu_usage = cpu_usage['cpu_usage'].mean()
avg_base_cpu_usage = base_cpu_usage['cpu_usage'].mean()
print(avg_cpu_usage / avg_base_cpu_usage)
plt.rc('font', size=10)
sns.barplot(x=['Base acquired CPU resources', 'Stratus acquired CPU resources'], y=[avg_base_cpu_usage, avg_cpu_usage])
plt.ylabel('Acquired CPU resources (Unit)')
```

```python
avg_cpu_utilization = cpu_usage['cpu_utilization'].mean()
avg_base_cpu_utilization = base_cpu_usage['cpu_utilization'].mean()
plt.rc('font', size=10)
sns.barplot(x=['Base CPU utilization', 'Stratus CPU utilization'], y=[avg_base_cpu_utilization, avg_cpu_utilization])
plt.ylabel('Average CPU utilization (%)')
```

```python
avg_memory_usage = memory_usage['memory_usage'].mean()
avg_base_memory_usage = base_memory_usage['memory_usage'].mean()
plt.rc('font', size=10)
sns.barplot(x=['Base acquired memory resources', 'Stratus acquired memory resources'], y=[avg_base_memory_usage, avg_memory_usage])
plt.ylabel('Acquired memory resources (Unit)')
```

```python
avg_memory_utilization = memory_usage['memory_utilization'].mean()
avg_base_memory_utilization = base_memory_usage['memory_utilization'].mean()
plt.rc('font', size=10)
sns.barplot(x=['Base memory utilization', 'Stratus memory utilization'], y=[avg_base_memory_utilization, avg_memory_utilization])
plt.ylabel('Average memory utilization (%)')
```
