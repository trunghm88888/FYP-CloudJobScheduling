```python
from collections import Counter

# count the occurrences of each point
c = Counter(zip(first_1000000[TASK_CPU_REQ_NAME],first_1000000[TASK_RUNTIME_NAME]))
# create a list of the sizes, here multiplied by 10 for scale
s = [10*c[(xx,yy)] for xx,yy in zip(first_1000000[TASK_CPU_REQ_NAME],first_1000000[TASK_RUNTIME_NAME])]


plt.scatter(first_1000000[TASK_CPU_REQ_NAME], first_1000000[TASK_RUNTIME_NAME], s=s)
plt.xlabel(TASK_CPU_REQ_NAME)
plt.ylabel(TASK_RUNTIME_NAME)
plt.show()
```
