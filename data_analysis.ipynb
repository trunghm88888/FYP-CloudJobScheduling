```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
from config_read_exports import *
```

```python
machine_df = pd.read_csv(machine_path)
```

```python
machine_df[['CPUs', 'Memory']].value_counts()
```

```python
cpu_count = machine_df['CPUs'].value_counts()
plt.figure(figsize=(10, 5))
sns.barplot(x=cpu_count.index, y=cpu_count.values)
plt.yscale('log')
plt.yticks([1, 10, 100, 1000, 10000], ['1', '10', '100', '1000', '10000'])
plt.ylabel('Number of machines')
plt.title('Machine CPU unit distribution')
```

```python
memory_count = machine_df['Memory'].value_counts()
plt.figure(figsize=(10, 5))
sns.barplot(x=memory_count.index, y=memory_count.values)
plt.yscale('log')
plt.yticks([1, 10, 100, 1000, 10000], ['1', '10', '100', '1000', '10000'])
plt.ylabel('Number of machines')
plt.title('Machine memory unit distribution')
```

```python
task_df = pd.read_csv(task_path).head(10_000_000)
```

```python
task_df['runtime'].describe()
```

```python
print(task_df['runtime'].quantile(0.95))
print(task_df['runtime'].quantile(0.99))
```

```python
sns.histplot(task_df['runtime'], bins=25)
plt.yscale('log')
plt.ylabel('Number of tasks')
plt.xlabel('Runtime (seconds)')
plt.title('Task runtime distribution')
plt.show()
```

```python
sns.histplot(task_df['start_time'] / 86400, bins=54)
plt.ylabel('Number of tasks')
plt.xlabel('Start time (days from the beginning of the trace)')
plt.title('Task start time distribution')
plt.xticks(np.arange(0, 18, 1))
plt.show()
```

```python
sns.histplot(task_df['CPU_request'], bins=10)
plt.yscale('log')
plt.ylabel('Number of tasks')
plt.xlabel('CPU request')
plt.title('Task CPU request distribution')
```

```python
sns.histplot(task_df['memory_request'], bins=5)
plt.yscale('log')
plt.ylabel('Number of tasks')
plt.xlabel('Memory request')
plt.title('Task memory request distribution')
plt.show()
```
